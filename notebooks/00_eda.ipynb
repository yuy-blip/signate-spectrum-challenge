# 00 - Exploratory Data Analysis

This notebook is for exploring the NIR spectral dataset.

## Setup
Place `train.csv` and `test.csv` in `data/raw/` before running.

```python
import sys
sys.path.insert(0, "../src")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from spectral_challenge.paths import DATA_RAW
from spectral_challenge.config import Config
```

```python
cfg = Config()
df_train = pd.read_csv(DATA_RAW / cfg.train_file)
df_test = pd.read_csv(DATA_RAW / cfg.test_file)
print("Train shape:", df_train.shape)
print("Test shape:", df_test.shape)
df_train.head()
```

```python
# Target distribution
fig, ax = plt.subplots(1, 1, figsize=(8, 4))
ax.hist(df_train[cfg.target_col], bins=30, edgecolor="k")
ax.set_xlabel(cfg.target_col)
ax.set_title("Target distribution")
plt.tight_layout()
plt.show()
```

```python
# Plot a few spectra
feat_cols = [c for c in df_train.columns if c not in {cfg.id_col, cfg.target_col}]
fig, ax = plt.subplots(1, 1, figsize=(12, 5))
for i in range(min(20, len(df_train))):
    ax.plot(df_train.loc[i, feat_cols].values, alpha=0.5)
ax.set_xlabel("Wavelength index")
ax.set_ylabel("Absorbance")
ax.set_title("Raw spectra (first 20 samples)")
plt.tight_layout()
plt.show()
```
